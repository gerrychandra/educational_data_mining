```python
import pandas as pd 
import numpy as np
import pickle

import matplotlib.pyplot as plt 
import seaborn as sns 

from src.evaluate import plot_metrics_cutoff, plot_confusion_matrix, plot_auc_roc, get_metric_score
```

Below is the performance evaluation of each model.

# Logistic Regression

```python
log_eval_data, log_eval_train, log_eval_test = pickle.load(open(in_path+'log_eval_noss.pkl', 'rb'))
```

By plotting `Probability vs. Various Metrics` curve, we can find the optimum `probability cutoff` that can make the `accuracy` and `recall` balanced. The most optimum `probability cutoff` for Logistic Regression is around **0.47**.

```python
log_cutoff_metrics = plot_metrics_cutoff(log_eval_test)
```

The `confusion matrix` is plotted with the optimum threshold probability shown in the `Probability vs. Various Metrics` curve above. With the threshold of **0.47**, `Logistic Regression` model results in `accuracy` of **0.65** and `recall` of **0.67** for train set.

The `AUC` of `Logistic Regression` model results in **0.73** for train set and **0.71** for test set.

```python
prob_thr = 0.47

fig, ax = plt.subplots(1,3, figsize=(17,5))
plot_confusion_matrix(ax[0], log_eval_data, 'TRAIN', thr=prob_thr)
plot_confusion_matrix(ax[1], log_eval_data, 'TEST', thr=prob_thr)
plot_auc_roc(ax[2], 
         log_eval_train.actual,
         log_eval_train.proba,
         log_eval_test.actual, 
         log_eval_test.proba)
```

# Random Forest

```python
RF_eval_data, RF_eval_train, RF_eval_test = pickle.load(open(in_path+'RF_eval_noss.pkl', 'rb'))
```

From the `Probability vs. Various Metrics` curve below, the optimum `probability cutoff` for `Random Forest` model is around 0.51. 

```python
RF_cutoff_metrics = plot_metrics_cutoff(RF_eval_test)
```

With the `probability threshold` of **0.51**, `Random Forest` model has similar metrics result compared with `Logistic Regression` model, which are **0.65** `accuracy` and **0.67** `recall`.

The `AUC` of `Random Forest` shows the same result as `Logistic Regression` for test set, **0.71**. But, `AUC` for train set shows a slightly higher result, which is **0.79**.

```python
prob_thr = 0.51

fig, ax = plt.subplots(1,3, figsize=(17,5))
plot_confusion_matrix(ax[0], RF_eval_data, 'TRAIN', thr=prob_thr)
plot_confusion_matrix(ax[1], RF_eval_data, 'TEST', thr=prob_thr)
plot_auc_roc(ax[2], 
         RF_eval_train.actual,
         RF_eval_train.proba,
         RF_eval_test.actual, 
         RF_eval_test.proba)
```

# Support Vector Machines

```python
SVC_eval_data, SVC_eval_train, SVC_eval_test = pickle.load(open(in_path+'SVC_eval_noss.pkl', 'rb'))
```

`Suppor Vector Machines` shows different characteristics compared to `Logistic Regression` and `Random Forest`. The optimum `probability cutoff` is so small, around 0.03.

```python
SVC_cutoff_metrics = plot_metrics_cutoff(SVC_eval_test)
```

The `SVM` results in **0.64** `accuracy` and **0.66** `recall` for test set. However, the metrics result for train set shows a significant difference, the `recall` for train set shows perfect **1.00** score.

The `AUC` also shows significant difference between train and test set. Train set gives perfect `AUC` score, while the test set only gives **0.66**.

```python
prob_thr = 0.0293

fig, ax = plt.subplots(1,3, figsize=(17,5))
plot_confusion_matrix(ax[0], SVC_eval_data, 'TRAIN', thr=prob_thr)
plot_confusion_matrix(ax[1], SVC_eval_data, 'TEST', thr=prob_thr)
plot_auc_roc(ax[2], 
         SVC_eval_train.actual,
         SVC_eval_train.proba,
         SVC_eval_test.actual, 
         SVC_eval_test.proba)
```

### Evaluation Summary
The `Logistic Regression` and `Random Forest` model shows a similar evaluation result while `SVM` model shows a very different characteristic. This might happen due to the fact that both `Logistic Regression` and `Random Forest` are **probabilistic** model and `SVM` itself is **non-probabilistic** model.

`SVM` shows a risk of overfitting since the performance in train set is so high compared to the test set. Therefore, `SVM` is not going to be used as the final model for production. 

Comparing `Logisitic Regression` and `Random Forest` is quite difficult since both gives similar results. However, `Logistic Regression` seems to have a better generalization capability compared to the `Random Forest` since the performance in the train set and test set is closer to each other. 

Due to the reasons above, `Logistic Regression` model is chosen to be the final model used for production when building the app.
